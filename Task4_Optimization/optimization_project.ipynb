```python
# ============================================
# TASK 4: OPTIMIZATION TECHNIQUE
# PRODUCTION OPTIMIZATION SYSTEM
# ============================================

# Step 1: Import libraries

import numpy as np
import matplotlib.pyplot as plt

from scipy.optimize import linprog
```

```python
# Step 3: Objective function

# Since linprog performs minimization,
# convert maximization to minimization

c = [-50, -40]

print("Objective Function:")
print(c)
```

```python
# ============================================
# CONSTRAINTS
# ============================================

# Step 4: Resource constraints

# 2A + B <= 100
# A + B <= 80

A = [
    [2, 1],
    [1, 1]
]

b = [100, 80]

print("Constraint Matrix:")
print(A)

print("\nConstraint Limits:")
print(b)
```

```python
# Step 5: Product quantities cannot be negative

x_bounds = (0, None)

print("Variable Bounds Set!")
```

```python
# Step 6: Solve optimization problem

result = linprog(
    c,
    A_ub=A,
    b_ub=b,
    bounds=[x_bounds, x_bounds],
    method='highs'
)

print("Optimization Completed!")
```

```python
# Step 7: Display optimal solution

print("\nOptimal Production:")

print("Product A =", result.x[0])

print("Product B =", result.x[1])

print("\nMaximum Profit =",
      -result.fun)
```
